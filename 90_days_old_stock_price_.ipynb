```python
def stockPrice(stock_symbol):
    
    #Master file
    writer = ExcelWriter('Sales_90_days.xls')
    
    for symbol  in stock_symbol:
        
        # Equity 
        series = "EQ"
        
        #Start date
        start_date = "12-07-2021"
        
        #End date
        end_date ="10-10-2021"
        
        #Calling equity history to fetch the data from a given range
        equity = equity_history(symbol,series,start_date,end_date)
        
        #checker if no data is available
        if len(equity.columns)==0:
            continue

        #sorting the date as it is in descending order
        equity=equity.sort_values(by = ['CH_TIMESTAMP'])
        
        #Selecting date and closing price of stocks
        closing_price = equity[['CH_TIMESTAMP','CH_CLOSING_PRICE']]
        
        # renaming columns
        closing_price.rename(columns = {'CH_TIMESTAMP':'DATE',
                                       'CH_CLOSING_PRICE':'CLOSING_PRICE'}
                                       ,inplace=True)
        
        #converting datframe to excel file along with adding sheets into existing files 
        closing_price.to_excel(writer ,sheet_name = f'{symbol}.NS',index = False)
    
    #saving 
    writer.save()
    
stockPrice(stock_symbol)
```
